### ILOC: nos permite seleccionar datos usando √≠ndices num√©ricos, como si fueran posiciones en una lista.


```python
###df.iloc[fila, columna]
"""donde:
fila ‚Üí posici√≥n o rango de filas (√≠ndice num√©rico).
columna ‚Üí posici√≥n o rango de columnas (tambi√©n por n√∫mero).
"""

print(df.iloc[2])        # Devuelve la fila en posici√≥n 2 (tercera fila, porque empieza en 0)

print(df.iloc[2, 1])     # Fila 2, columna 1

print(df.iloc[0:5])      # Filas desde la 0 hasta la 4

print(df.iloc[[0, 2, 4]])  # Filas 0, 2 y 4 dentro de un corchete en el segundo corchete no hay , para columnas

print(df.iloc[:, 0:3])   # Todas las filas, columnas 0 a 2 Subconjunto filas + columnas

print(df.iloc[0:3, 1:4]) # Filas 0 a 2, columnas 1 a 3
```

```python
# Acceder a la fila n√∫mero 2 (√≠ndice 2)
print("Registros de la fila n√∫mero 2:\n", df.iloc[2])  

# Acceder a la celda en la fila 2, columna 1 (Beatriz, 22 a√±os)
print("Registro de la fila 2, columna 1:",df.iloc[2, 1])

# Acceder a las filas de 1 a 3 (excluye la ultima)
print("Registros de las filas 1 a 2:",df.iloc[1:3]) 
```

###  <font color='green'>Crear nuevas columnas</font>

Recordando los diccionarios de Python, cuando agregamos nuevas claves lo hacemos de la siguiente forma:

```python
# Creamos un diccionario con un solo par clave-valor, que es una lista en nuestro caso.
diccionario_dias = {"dias_laborales": ["Lunes", "Martes", "Mi√©rcoles", "Jueves", "Viernes"]}

print(diccionario_dias)

# Asignamos un nuevo par clave-valor.
diccionario_dias["dias_no_laborales"] =  ["S√°bado", "Domingo"]

print(diccionario_dias)
```

```python
df_dias = pd.DataFrame({"dias_laborales": ["Lunes", "Martes", "Mi√©rcoles", "Jueves", "Viernes"]})
df_dias
```

```python
df_dias["dias_no_laborales"] = ["S√°bado", "Domingo", "", "", ""]
df_dias
```

## <font color='green'>Combinaci√≥n de DataFrames</font>

Podemos combinar dos DataFrames en un proceso similar a la operaci√≥n JOIN, de tablas en SQL.

Con Python, utilizamos **pd.concat()**, que recibe una lista con DataFrames que deben tener las mismas columnas.

```python
df1 = pd.DataFrame({"mes": "enero", "ventas":np.random.randint(1,100, 5)}) ##5 enteros entre 1 y 99.
df1
```

```python
df2 = pd.DataFrame({"mes": "febrero", "ventas":np.random.randint(1,100, 5)})
df2
```

```python
df3 = pd.concat([df1, df2])
df3
```

### <font color='green'>Reindexando</font>

Podemos observar que los indices quedaron duplicados y desordenados, por lo que para evitar confusiones se recomienda reiniciar el indice de las filas luego de hacer Uniones de este estilo.

```python
df3 = df3.reset_index(drop=True) #¬øQu√© ocurre si no ponemos "drop=True" (que es el valor por defecto)?
df3
```

```python
df33 = df3.reset_index(drop=False) 
df33
```

```python
df4 = pd.DataFrame({"mes": "marzo", "ventas": np.random.randint(1,100, 1)})
df4
```

Para simplificar, utilizamos el parametro **ignore_index** del m√©todo **concat** al realizar la union de DataFrames. De esta manera, el indice se reiniciar√° autom√°ticamente y no se duplicara.

```python
df5 = pd.concat([df3, df4], ignore_index = True)
df5
```

## <font color='green'>Indexaci√≥n de un DataFrame</font>

Indexar un DataFrame es una operaci√≥n fundamental en Pandas, que nos permite seleccionar un conjunto de datos espec√≠fico del DataFrame. La indexaci√≥n permite filtrar, ordenar y manipular los datos de una manera m√°s f√°cil y r√°pida. Distinguiremos esencialmente tres indexaciones fundamentales:

- **Indexaci√≥n por posici√≥n:** seleccionar una o varias filas y/o columnas del DataFrame mediante sus √≠ndices num√©ricos.
- **Indexaci√≥n por etiquetas:** seleccionar una o varias filas y/o columnas del DataFrame mediante sus etiquetas o nombres.
- **Indexaci√≥n booleana:** seleccionar filas del DataFrame que cumplan una o varias condiciones l√≥gicas.

De estas tres, la **indexaci√≥n booleana** puede resultar m√°s c√≥moda para el analisis de datos ya que permite hacer filtros basados en condiciones aplicados a la **Data** en lugar de a los **Indices**.

### <font color='green'>Indexaci√≥n por posiciones</font>

### <font color='green'>Indexaci√≥n booleana</font>

Podemos seleccionar datos poniendo condiciones, que podemos combinar utilizando los operadores booleanos & (and) y | (or). Notemos que las condiciones deben ponerse entre par√©ntesis, cuando haya m√°s de una.

```python
df[(df['edad'] >= 30) & (df['anos_experiencia'] >=5)]
```

### <font color='green'>El m√©todo query()</font>

El m√©todo llamado **query()** nos permite realizar una indexaci√≥n booleana m√°s sencilla:

```python
df.query("edad <= 21 or anos_experiencia == 0")
```

Como puedes ver:

- No son necesarios los parentesis **().**
- Podemos utilizar **"and"** y **"or"** en vez de **"&"** y **"|"**.

## <font color='green'>Exportando un DataFrame</font>

Podemos exportar f√°cilmente un DataFrame a un archivo csv (u otro formato, como xls) mediante el m√©todo **to_csv**. Su uso es el siguiente:

```python
df.query.to_csv("nombre.csv", index = False, sep = ',', header = False, columns = ['edad','evaluacion_adaptativa'])
```

los par√°metros de este m√©todo son:

  - **nombre.cvs:** corresponde al nombre del archivo que exportaremos (quedar√° en la misma carpeta que nuestro archivo de Jupyter o Python.
  - **index:** agrega o no una columna adicional de √≠ndice. Por defecto es True, por lo que denbemos especificar si no la queremos.
  - **sep:** corresponde al separador de datos. En nuestro caso, ser√°n comas.
  - **header:** similar a index, agrega o no la cabecera.
  - **columns**: lista de columnas que queremos que se exporten (si no precisamos, se exportar√°n todas). Podemos indicar las columnas utilizando sus etiquetas (como en nuestro ejemplo) o utilizando posiciones. En tal caso usamos columns=df.columns[[0, 2]]
