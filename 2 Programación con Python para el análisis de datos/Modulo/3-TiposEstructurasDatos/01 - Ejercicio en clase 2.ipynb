Un **DataFrame** de **Pandas** tambien puede ser construido a partir de estructuras de datos mas simples de Python, como las **listas** y los **diccionarios**, veamos algunos ejemplos:

```python
import pandas as pd

# Una lista de diccionarios.

lista_de_diccionarios = [

    {'nombre': 'Ana', 'profesion': 'guitarra', 'edad': '34'},
    {'nombre': 'Luis', 'profesion': 'fot√≥grafo', 'edad': '27'},
    {'nombre': 'Mar√≠a', 'profesion': 'profesora', 'edad': '42'},
    {'nombre': 'Pedro', 'profesion': 'chef', 'edad': '39'},
    {'nombre': 'Isabel', 'profesion': 'arquitecta', 'edad': '29'},
    {'nombre': 'Sergio', 'profesion': 'm√∫sico', 'edad': '25'},
    {'nombre': 'Carla', 'profesion': 'ingeniera', 'edad': '31'},
    {'nombre': 'David', 'profesion': 'estudiante', 'edad': '20'},
    {'nombre': 'Luc√≠a', 'profesion': 'abogada', 'edad': '36'},
    {'nombre': 'Jorge', 'profesion': 'actor', 'edad': '28'}

]

df = pd.DataFrame(lista_de_diccionarios)
df
```

```python
# Una lista de listas. (listas anidadas):

lista_anidada = [

    ['Ana', 'guitarra', '34'],
    ['Luis', 'fot√≥grafo', '27'],
    ['Mar√≠a', 'profesora', '42'],
    ['Pedro', 'chef', '39'],
    ['Isabel', 'arquitecta', '29'],
    ['Sergio', 'm√∫sico', '25'],
    ['Carla', 'ingeniera', '31'],
    ['David', 'estudiante', '20'],
    ['Luc√≠a', 'abogada', '36'],
    ['Jorge', 'actor', '28']
]


df = pd.DataFrame(lista_anidada, columns = ['nombre', 'profesion', 'edad'])
df
```

```python
# Obtener la primera fila.
df.head(1)
```
