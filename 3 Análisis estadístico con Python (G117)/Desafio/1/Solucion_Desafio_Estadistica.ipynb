### Ejercicio 1:

Se solicita calcular el promedio general de los salarios, desviación estándar, quintiles y rango utilizando la columna `salary_in_usd` para asegurar una comparación estandarizada.

```python
import pandas as pd

# Cargar datos
df = pd.read_csv("ds_salaries.csv")

# Indicadores para salary_in_usd
mean_salary = df['salary_in_usd'].mean()
std_salary = df['salary_in_usd'].std()
quintiles = df['salary_in_usd'].quantile([0.2, 0.4, 0.6, 0.8])
salary_range = df['salary_in_usd'].max() - df['salary_in_usd'].min()

print(f"Promedio: ${mean_salary:,.2f} USD")
print(f"Desviación Estándar: ${std_salary:,.2f} USD")
print("Quintiles:")
print(quintiles)
print(f"Rango: ${salary_range:,.2f} USD")
```

**Interpretación Ejercicio 1:**
- El **promedio** (~$137,570) nos da una idea del centro de la distribución, pero la **desviación estándar** (~$63,055) es bastante alta, lo que indica que los salarios varían significativamente respecto al promedio.
- El **rango** de más de $444k confirma la existencia de una gran brecha entre los salarios iniciales y los de niveles ejecutivos.
- Los **quintiles** nos permiten segmentar a la población: el 20% que menos gana percibe menos de $84k, mientras que el 20% superior supera los $185k.

### Ejercicio 2:

Se realiza una comparación entre los salarios agrupados por 3 categorías: `experience_level`, `company_size` y `employment_type`.

```python
categories = ['experience_level', 'company_size', 'employment_type']

for cat in categories:
    print(f"\n--- Análisis por {cat} ---")
    stats = df.groupby(cat)['salary_in_usd'].agg(['mean', 'median', 'std', 'count'])
    # Calculamos Coeficiente de Variación (CV) para medir representatividad relativa
    stats['cv'] = stats['std'] / stats['mean']
    print(stats)
```

1. **¿Para cuál categoría las medidas son más representativas?**
Para el nivel de experiencia **'EX' (Executive)** y **'SE' (Senior)**. 
- Tienen los Coeficientes de Variación (CV) más bajos (0.36 y 0.37 respectivamente).
- Esto significa que los datos están más agrupados cerca del promedio y hay menos dispersión relativa.

2. **¿Para cuál son menos representativas?**
Para el tipo de contrato **'CT' (Contract)** y **'PT' (Part-time)**.
- Sus CV son muy elevados (> 0.9), lo que indica una dispersión extrema.
- Además, el tamaño de la muestra es muy pequeño (10 y 17 registros), lo que hace que cualquier valor atípico mueva mucho el promedio, restándole confianza.
