```python
import pandas as pd

df = pd.read_csv("worldcup2014.csv")

# a. pertenecer a África
total_paises = len(df)
paises_africa = len(df[df['continent'] == 'africa'])
prob_africa = paises_africa / total_paises

print(f"Probabilidad de pertenecer a África: {prob_africa}")
```
