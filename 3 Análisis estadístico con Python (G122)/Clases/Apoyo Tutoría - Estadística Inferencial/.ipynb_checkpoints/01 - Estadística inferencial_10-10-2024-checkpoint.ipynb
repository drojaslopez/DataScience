## <font color="green">Prueba de hipótesis</font>

Python nos permite determinar fácilmente valores críticos e implementar modelos para pruebas de hipótesis. Veremos a continuación algunos de estos procedimientos para aplicar pruebas de hipótesis sobre una base de datos considerándola **una muestra de una población total.**


## Ideas fundamentales:

* 1 La evaluación de una hipotesis
* 2 La distribución normal de los datos
* 3 Comprender que es el valor P
* 4 Comprender qué es el valor t

### 1. Evaluación de una hipótesis (EJEMPLO)

**Ejemplo sencillo:** Imaginemos que alguien afirma que "el promedio de la altura de los estudiantes de una clase es mayor a 1.70 metros". Esta afirmación es una hipótesis.

**Hipótesis nula (H₀):** Es la posición inicial que asume que no hay un efecto significativo o cambio esperado.En este caso, sería: **"El promedio de la altura (𝜇) es menor o igual a 1.70 metros."**

**Hipótesis alternativa (H₁):** Es lo que queremos demostrar. Aquí sería: "El promedio de la altura es mayor a 1.70 metros."

La evaluación de una hipótesis es el proceso de recoger datos (las alturas de los estudiantes) y hacer cálculos estadísticos para determinar si hay suficiente evidencia para rechazar la hipótesis nula (H₀), es decir, que la altura promedio no es mayor a 1.70 mts. o bien **"El promedio de la altura (𝜇) es menor o igual a 1.70 metros."**

### 2. Distribución normal de los datos

**Ejemplo sencillo:** Si medimos las alturas de todos los estudiantes de una escuela, la mayoría de las alturas estarán cerca del promedio, y muy pocas serán extremadamente altas o extremadamente bajas. Si representamos esto en un gráfico, obtendremos una curva en forma de campana, llamada distribución normal.

Los datos que siguen una distribución normal tienen una forma simétrica y la mayoría de los valores están agrupados alrededor de la media.
La distribución normal es importante porque muchos fenómenos naturales (como la altura) tienden a seguir este patrón.



### "Valor t" y "Valor p"

**¿Qué es el valor t?**

El valor t es una medida que usamos para comparar los datos de un grupo con lo que esperaríamos encontrar si no hubiera diferencias reales. Es como un marcador de cuánto se desvía nuestra muestra del "promedio esperado".

Por ejemplo:
* Imagina que estás probando si un nuevo medicamento para el dolor funciona.
* Si no hay efecto, el promedio del dolor antes y después sería el mismo.
* El valor t mide qué tan lejos está la diferencia entre antes y después de lo que esperaríamos si el medicamento no funcionara.

Un valor t grande significa que la diferencia es más fuerte y probablemente no es por pura suerte.

**¿Qué es el valor p?**

El valor p nos dice la probabilidad de que esos resultados ocurran solo por casualidad.

Por ejemplo:
* Si el valor p es 0.03, significa que hay un 3% de probabilidad de que los resultados sean por pura suerte, asumiendo que el medicamento no funciona (hipótesis nula).
* Un valor p pequeño (por ejemplo, menor a 0.05) significa que hay menos probabilidad de que el resultado sea por casualidad, lo que nos da más confianza para decir: "El medicamento sí funciona".

### 3. Comprender qué es el valor P

**Ejemplo sencillo:** Volviendo al ejemplo de las alturas, el valor P nos dice qué tan probable es obtener los datos que tenemos si la hipótesis nula (H₀) es verdadera. **El promedio de la altura de los estudiantes es menor o igual a 1.70 metros.**

Si el valor P es muy pequeño (por ejemplo, menor a 0.05), significa que hay una probabilidad muy baja de que los datos observados ocurran por azar si H₀ es verdadera, y por lo tanto, podemos rechazar H₀.
Si el valor P es grande, significa que no hay suficiente evidencia para rechazar H₀.
En resumen, el valor P es una medida de la evidencia contra la hipótesis nula

### Resultado posible

* Si los datos nos llevan a un valor p de 0.02 (menor a 0.05), concluimos:
"Con un nivel de confianza del 95%, podemos afirmar que el promedio de la altura de los estudiantes es mayor a 1.70 metros.", Entonces rechazamos la hipotesis nula, **ya que esta decia que la altura promedio era menor o igual a 1.70**

* Si el valor p resulta ser 0.10 (mayor a 0.05), concluimos:
"No hay suficiente evidencia para afirmar que el promedio de la altura de los estudiantes es mayor a 1.70 metros.", Entonces, **no rechazamos la hipótesis nula**, porque no tenemos suficiente evidencia para apoyar la hipótesis alternativa, la cual decía que el promedio de la altura es mayor a 1.70 metros.

### 4. Comprender qué es el valor t

El valor **𝑡** es un estadístico que mide cuántas unidades de error estándar hay entre la media muestral y el valor hipotético de 𝐻0.

**Ejemplo intuitivo:** Si alguien afirma que la altura promedio es 1.70 metros, y calculamos **𝑡 = 3**, significa que la media muestral está a 3 veces el error estándar de la media hipotética.

Cuanto más grande (en valor absoluto) sea **𝑡**, más evidencia hay contra **𝐻0**.

Un valor **𝑡 grande** lleva a un valor **𝑝 pequeño**, indicando que los datos no respaldan 𝐻0

### Distribución normal y evaluación de hipótesis

Al evaluar una hipótesis, a menudo necesitamos comparar nuestros datos con una distribución teórica para determinar si los resultados que observamos son inusuales. En muchos casos, se asume que los datos siguen una distribución normal.

La distribución normal está directamente relacionada con la evaluación de hipótesis y el cálculo del valor P, ya que muchos procedimientos estadísticos asumen que los datos siguen una distribución normal, especialmente cuando estamos trabajando con muestras grandes.

**Por ejemplo,** si estamos evaluando si las alturas de los estudiantes son mayores a 1.70 metros, podríamos suponer que las alturas siguen una distribución normal.
Luego calculamos la media y la desviación estándar de nuestra muestra y vemos cómo se comparan estos resultados con la distribución normal.


### Distribución normal y valor P

El valor P se basa en la probabilidad de observar un resultado tan extremo (o más extremo) que el obtenido, asumiendo que la hipótesis nula es verdadera. Para calcular el valor P, necesitamos saber cómo se comportan los datos si H₀ es cierta. Aquí es donde entra la distribución normal:

La distribución normal nos da una referencia de cómo deberían distribuirse los datos si no hubiera un efecto significativo.
Usamos la curva de la distribución normal para calcular el valor P. Si el valor observado (por ejemplo, la media de las alturas) está muy lejos del centro de la distribución (es decir, en las "colas" de la campana), el valor P será pequeño. Esto indica que es poco probable que nuestros datos hayan ocurrido bajo H₀.

**En resumen:**

Al suponer que los datos siguen una distribución normal, podemos calcular la probabilidad de observar los resultados que obtenemos, que es lo que nos permite calcular el valor P.
Si nuestros datos están muy lejos de lo esperado en una distribución normal, el valor P será pequeño, lo que indica que hay evidencia para rechazar la hipótesis nula.
Ejemplo visual:


Imagina la curva normal:

* Centro de la curva: lo que esperamos bajo la hipótesis nula.
* Colas de la curva: resultados extremos (improbables si H₀ es verdadera).
* Valor P: mide la probabilidad de caer en esas colas si H₀ es cierta.

Esto vincula la distribución normal con la decisión sobre la hipótesis, usando el valor P como

* mayor a 1.70 metros
* Menor a 1.70 metros
* es distinta  a 1.70

## El contexto de las hipótesis:

Estamos probando la hipótesis de que el promedio de la altura de los estudiantes es mayor a 1.70 metros. Las opciones en el parámetro alternative nos permiten definir cómo vamos a comparar nuestros datos contra el valor esperado.

## Opciones de alternative: 


### *alternative='greater':**

**Hipótesis nula (H₀):** El promedio de la altura es menor o igual a 1.70 metros.
 
**Hipótesis alternativa (H₁):** El promedio de la altura es mayor a 1.70 metros.

Aquí, sólo nos interesa si las alturas son mayores a 1.70, y por eso utilizamos **"greater"**. Si la prueba encuentra evidencia de que el promedio es mayor, rechazamos H₀.
Ejemplo: Queremos probar si los estudiantes son más altos que 1.70 metros. Si el resultado es significativo, podemos decir que el promedio es mayor a 1.70 metros.

### alternative='less':

**Hipótesis nula (H₀):** El promedio de la altura es mayor o igual a 1.70 metros.

**Hipótesis alternativa (H₁):** El promedio de la altura es menor a 1.70 metros.


Esta opción se usa cuando queremos ver si los estudiantes son más bajos que 1.70 metros. Estamos interesados solo en los resultados menores a 1.70, por lo que se usa **"less".**

Ejemplo: Si queremos probar que los estudiantes son más bajos que 1.70 metros, usamos alternative='less'.

### alternative='two-sided' (opción por defecto):

**Hipótesis nula (H₀):** El promedio de la altura es exactamente igual a 1.70 metros.

**Hipótesis alternativa (H₁):** El promedio de la altura es diferente a 1.70 metros (puede ser mayor o menor).

Aquí, estamos interesados en ambas direcciones, es decir, si las alturas son diferentes (mayores o menores) a 1.70 metros.
Ejemplo: Si queremos ver si el promedio es simplemente diferente de 1.70 (sin importar si es mayor o menor), usamos alternative=**'two-sided'.**

**Resumen:**


**greater:** Probamos si el valor de la muestra es mayor que el valor de referencia.

**less:** Probamos si el valor de la muestra es menor que el valor de referencia.

**two-sided:** Probamos si el valor de la muestra es simplemente diferente (mayor o menor) al valor de referencia.



En Python, usarías estos parámetros dependiendo de qué tipo de hipótesis estás evaluando, ajustando el código según lo que quieras comprobar sobre los dato

# Lo aplicamos en una funciòn de prueba_hipotesis_media

```python
import pandas as pd
import numpy as np
from scipy import stats

# Filtrar los datos para obtener solo los pesos de los pingüinos Adelia
adelia = pinguins[pinguins["species"] == "Adelie"]["body_mass_g"]
adelia = adelia.dropna()

# Función para la prueba de hipótesis
def prueba_hipotesis_media(datos, mu, colas):
    datos = datos.dropna()  # Eliminar valores NaN, por si acaso
    alpha=round(np.random.uniform(0.01, 0.1),2)
    confianza = 100 * (1 - alpha)
    
    t_stat, p_value = stats.ttest_1samp(datos, mu, alternative=colas)
    
    if p_value < alpha:
        if colas == 'two-sided':
            print(f"Con nivel de confianza de {confianza}% se rechaza la hipótesis nula: la media es distinta de {mu}.")
        elif colas == 'greater':
            print(f"Con nivel de confianza de {confianza}% se rechaza la hipótesis nula: la media es mayor que {mu}.")
        elif colas == 'less':
            print(f"Con nivel de confianza de {confianza}% se rechaza la hipótesis nula: la media es menor que {mu}.")
    else:
        print(f"Con nivel de confianza de {confianza}% no se puede rechazar la hipótesis nula: la media es igual a {mu}.")
    
    print(f"Valor t: {t_stat}")
    print(f"Valor p: {p_value}\n")

# Calcular la media de los pesos de los pingüinos Adelia
media = np.mean(adelia)
print(f"Media de los pesos de los pingüinos Adelia: {media}")

# Probar con diferentes valores de medias
medias = [media - 50, media - 1, media, media + 50, media + 3000]

# Lista de tipos de prueba
tipos_de_prueba = ['two-sided', 'greater', 'less']

# Realizar las pruebas de hipótesis para cada valor de media y para cada tipo de prueba
for mu in medias:
    for tipo in tipos_de_prueba:
        print(f"Prueba con media = {mu} y tipo de prueba '{tipo}':")
        prueba_hipotesis_media(adelia, mu, tipo)
```

### Analisis caso de media -50 (3650 grms)

**Prueba con media = 3650 y tipo de prueba 'two-sided': ESTAMOS DICIENDO QUE LA MEDIA ES DIFERENTE DE 3650**

* El valor t calculado es 1.34, y el valor p es 0.182.
    
Para interpretar los resultados, con un nivel de confianza del 95%, si el valor p es mayor que 0.05, no rechazamos la hipótesis nula.

El valor p de 0.182 es mayor que 0.05, lo que significa que no podemos rechazar la hipótesis nula. En otras palabras, con un nivel de confianza del 95%, no hay suficiente evidencia para decir que la media del peso de los pingüinos Adelia es diferente de 3650 gramos.

Aunque la media de tus datos es 3700.66 gramos, la diferencia no es estadísticamente significativa en este caso, lo que implica que, basándonos en este análisis, la media podría ser igual a 3650 gramos.

* El valor t calculado es 1.34, y el valor p es 0.182.
    
Ahora, para una prueba 'greater', si el valor p de la prueba es 0.182, aún no es suficientemente bajo como para rechazar la hipótesis nula. El valor p debe ser menor que 0.05 para que rechacemos 𝐻0 y afirmemos que la media es mayor que 3650 gramos.

Dado que el valor p es 0.182, no podemos rechazar la hipótesis nula, lo que significa que no hay suficiente evidencia para afirmar que la media es mayor que 3650 gramos.

Prueba con media = 3650 y tipo de prueba 'less': ESTAMOS DICIENDO QUE LA MEDIA ES MENOR QUE 3650**

* El valor t calculado es 1.34, y el valor p es 0.182.
    
En la prueba two-sided que realizamos previamente, el valor p fue 0.182. Para realizar la prueba 'less', necesitamos adaptar el valor p para que solo se considere la cola inferior (la probabilidad de obtener un valor t tan negativo como el que hemos calculado, si la media fuera realmente 3650).

En este caso, si mantuviéramos la misma distribución y ajustamos el valor p para una prueba 'less', la probabilidad sería la mitad de la prueba de dos colas, porque estamos considerando solo una cola (la izquierda). Entonces, sería:

### ¿Por qué no rechazamos 𝐻 0 en ninguno de los casos?

El valor p nos dice cuánta probabilidad hay de obtener los resultados observados si la hipótesis nula fuera verdadera. En este caso, con un valor p de 0.182 (en la prueba two-sided), 0.182 / 2 = 0.091 (en la prueba 'less'), y un valor p similar en la prueba 'greater', el valor p es siempre mayor que 0.05. Esto implica que:

* No tenemos suficiente evidencia para rechazar la hipótesis nula.

* La hipótesis nula establece que la media es igual a 3650 gramos, y con el valor p tan alto, no podemos decir que la media sea significativamente diferente de ese valor en ninguno de los tres casos.

### Analisis del ejemplo

El valor de t corresponde a la diferencia entre la media muestral $\overline{x}$ y la media hipotética $\mu$, normalizada dividiendo por el error estándar de la muestra (la desviación estándar de la misma). Dado que la distribución de probabilidad es simétrica, los valores correspondientes a "media-50" y "media+50" presentan valores t iguales en valor absoluto, y valores p iguales. Observamos también que el valor de t aumenta en valor absoluto mientras mayor sea la distancia de la media "hipotética" con la "media muestral", y cuando esto ocurre el valor de p disminuye cada vez más acercándose a cero.


¿Cómo interpretamos esto? Mientras mayor sea la diferencia entre la media hipotética y la media muestral, es cada vez menos probable que la media hipotética sea, efectivamente, la media de la población. ¿Qué probabilidad estamos dispuestos a aceptar? Vamos a **rechazar** la hipótesis nula si el valor de p es menor que el nivel de significancia escogido. Así, en nuestro ejemplo, si asumimos un nivel de significancia igual a 0.1, por ejemplo, tendremos que:

*  No podemos rechazar la hipótesis de que la media poblacional sea "media-50", ya que p = 0.18, que es mayor que 0.1
*  Si podemos rechazar la hipótesis de que la media poblacional sea "media+3000", ya que p = 3.136232388747058e-125, que es menor que 0.1

Podemos observar que nuestra hipótesis alternativa sería "de dos colas", ya que estamos afirmando que la media sería diferente a la supuesta pero no indicamos en qué sentido. Python nos permite considerar esto agregando un parámetro a la función

```python
t_stat, p_value = st.ttest_1samp(adelia, media-1000, alternative='greater')
print(f"Valor t: {t_stat}")
print(f"Valor p: {p_value}\n")

t_stat, p_value = st.ttest_1samp(adelia, media-100, alternative='greater')
print(f"Valor t: {t_stat}")
print(f"Valor p: {p_value}\n")

t_stat, p_value = st.ttest_1samp(adelia, media-1, alternative='greater')
print(f"Valor t: {t_stat}")
print(f"Valor p: {p_value}\n")


t_stat, p_value = st.ttest_1samp(adelia, media+1, alternative='greater')
print(f"Valor t: {t_stat}")
print(f"Valor p: {p_value}\n")

t_stat, p_value = st.ttest_1samp(adelia, media+10, alternative='greater')
print(f"Valor t: {t_stat}")
print(f"Valor p: {p_value}\n")

t_stat, p_value = st.ttest_1samp(adelia, media+100, alternative='greater')
print(f"Valor t: {t_stat}")
print(f"Valor p: {p_value}\n")

t_stat, p_value = st.ttest_1samp(adelia, media+1000, alternative='greater')
print(f"Valor t: {t_stat}")
print(f"Valor p: {p_value}\n")
```

"Alternative='greater' indica que nuestra hipótesis alternativa es que la media de la población es mayor que la que estamos suponiendo. Si consideramos un valor **muy menor que** la media poblacional como media hipotética, es muy probable que debamos rechazar dicha hipótesis nula a cambio de la alternativa, que postula que la media es **mayor** que la hipotética. Naturalmente, mientras más nos vayamos acercando a la media real el valor de p irá aumentando. Notarenos que, aun si nuestra media muestral propuesta es menor que la muestral, no podemos descartar la hipótesis nula pues sigue teniendo una alta probabilidad.

¿Qué ocurre cuando comenzamos a tomar valores **mayores que** nuestra media muestral? El valor de p irá aumentando (hasta hacerse 1, o muy cerca de 1). En el último caso, en que hemos tomado una media hipotética igual a "media + 1000", obtenemos un valor de p igual a 1 que nos indicaría... **¡qué no podemos rechazar la hipótesis!** ¿Significa que debemos aceptar, realmente, que esta es la media? En realidad no. Lo que estamos afirmando que la hipótesis alternativa (que la media poblacional sea **mayor que "media+1000"**) es muuuucho menos probable.

Tenemos otras posibilidades, además: "alternative=less" (menor que) y "alternative=two-side" (dos lados) que es la opción por defecto.

## <font color="green">Infiriendo sobre la media - Intervalos de confianza</font>

Considerando la misma muestra anterior, podemos construir un intervalo de confianza para estimar la media poblacional. Utilizaremos un nivel de confianza del 95%.


Un intervalo de confianza es un rango de valores en el que se espera que se encuentre la verdadera media poblacional, con un nivel de certeza determinado (en este caso, el 95%).

* Nivel de confianza del 95% significa que si repitiéramos el muestreo muchas veces, en el 95% de las veces el intervalo calculado incluiría la verdadera media de la población.

* El intervalo de confianza se calcula usando la media de la muestra y un margen de error basado en la desviación estándar y el tamaño de la muestra.


El intervalo de confianza te dará un rango, por ejemplo, "la media de la población está entre 3600 y 3800 gramos con un 95% de confianza."



### Con distribución t

```python
# crear un intervalo de confianza del 95% para el peso medio de la población
st.t.interval (confidence= 0.95, df = len (adelia) -1, loc = media, scale = st.sem (adelia))
```

Podemos identificar aquí los siguientes parámetros:

*  Confidence=0.95: corresponde al nivel de confianza. Observa que se da directamente.

*  df=len(data)-1: corresponde al número de grados de libertad para la distribución t. Se calcula restando 1 al número de observaciones en la muestra.

*  loc=media: utilizamos la media de la muestra como estimador para nuestro intervalo de confianza.

*  scale=st.sem(adelia): Se calcula el error estándar de la media (SEM) de la muestra de datos utilizando st.sem(adelia). El error estándar de la media es una medida de la variabilidad de la media muestral y se utiliza para calcular la amplitud del intervalo de confianza.

**SEM (Standard Error of the Mean o Error Estándar de la Media)**

Calculando el error y definiendo la confiabilidad podemos escribir una afirmación en lenguaje natural:

```python
alpha=0.95
# crear un intervalo de confianza del 95% para el peso medio de la población
a,b=st.t.interval (confidence= alpha, df = len (adelia) -1, loc = media, scale = st.sem (adelia))
error=round(b-media,2)
media_estimada=round(media,2)
conf=alpha*100
print(f'La media poblacional es {media_estimada}, con un margen de error de {error} y confiabilidad del {conf}%')
```

## Interpretación del intervalo de confianza:

El intervalo de confianza del 95% que calculamos va desde 3626.92 grs hasta 3774.39 grs. Esto significa que, con un 95% de confianza, el peso medio de la población se encuentra dentro de ese rango.

## ¿Qué pasa si el promedio cae fuera de ese rango?


 * Si realizas otro análisis o tomas otra muestra, y el promedio observado en esa nueva muestra cae por debajo de 3626.92 o por encima de 3774.39, se podría rechazar la hipótesis de que el verdadero promedio de la población está dentro de ese rango.
 
 
 
 * En otras palabras, si tu media cae fuera de ese intervalo, es poco probable (menos del 5%) que el promedio verdadero esté dentro de ese rango, lo cual te llevaría a rechazar la hipótesis nula (la suposición de que el promedio está en ese rango).

## Relación con la confiabilidad:


Al utilizar un nivel de confianza del 95% (α = 0.05), estás diciendo que 95 veces de cada 100 esperas que el intervalo contenga la verdadera media poblacional. Pero hay un 5% de probabilidad de error de que tu intervalo no contenga la verdadera media.

## Ejemplo de afirmación:

"La media poblacional es 3700.66, con un margen de error de 73.74 y una confiabilidad del 95%. Esto significa que estamos 95% seguros de que la media poblacional verdadera se encuentra entre 3626.92 y 3774.39. Si se encuentra fuera de este rango, rechazamos la hipótesis de que el promedio poblacional está dentro de este intervalo.

## 1. Distribución t vs. Distribución normal

**Distribución t: Se utiliza principalmente cuando:**

La muestra es pequeña (generalmente n < 30).
La desviación estándar de la población es desconocida, y se utiliza la desviación estándar de la muestra como estimador.
La distribución de los datos es aproximadamente normal.

**Distribución normal: Se utiliza cuando:**

La muestra es grande (generalmente n ≥ 30), lo que permite aplicar el Teorema del Límite Central.
La desviación estándar de la población es conocida o asumimos que se distribuye normalmente.
Los datos se distribuyen normalmente.

2. Cálculo de los intervalos de confianza
**Usando st.t.interval:**


```python
st.t.interval(confidence=0.95, df=len(adelia)-1, loc=media, scale=st.sem(adelia))
df=len(adelia)-1 #Grados de libertad.
scale=st.sem(adelia) #Error estándar de la media basado en la desviación estándar de la muestra.
```

**Usando st.norm.interval:**

**st.norm.interval(alpha, loc=media, scale=np.std(adelia))**

* Aquí no se utilizan grados de libertad.

* scale=np.std(adelia): La desviación estándar de la muestra se utiliza directamente, sin ajuste por el tamaño de la muestra. Esto es menos adecuado para muestras pequeñas.

### Con distribución normal




```python
# crear un intervalo de confianza del 95% para el peso medio de la población
st.norm.interval(alpha, loc=media, scale=np.std(adelia))
```

La diferencia entre el cálculo de un intervalo de confianza utilizando la distribución t y la distribución normal radica en cómo se asume que se distribuyen los datos y en el tamaño de la muestra. 

## 3. Resultados del intervalo de confianza

**Intervalo con distribución t:** Para tus datos, el intervalo de confianza del 95% resultó ser aproximadamente **(3626.92, 3774.39)**. Este intervalo es más estrecho y tiene en cuenta la variabilidad de la muestra con el ajuste correspondiente.

**Intervalo con distribución normal:** El intervalo de confianza del 95% usando la distribución normal resultó ser **(2804.87, 4596.45)**. Este intervalo es más amplio porque no ajusta adecuadamente por el tamaño de la muestra y usa la desviación estándar total de los datos.

**Conclusión:**

**Tamaño de la muestra:** La distribución t es más adecuada para muestras pequeñas porque ofrece un intervalo de confianza más ajustado, reflejando la incertidumbre adicional.

**Distribución normal:** Es más amplia y asume que la variabilidad de la muestra es representativa de la población, lo cual puede no ser cierto si la muestra es pequeña.

si tenemos una muestra pequeña, es recomendable utilizar la distribución t para obtener un intervalo de confianza más preciso. Si la muestra es lo suficientemente grande (n ≥ 30), podemos usar la distribución normal, ya que la estimación de la media poblacional será más robusta.

## <font color="green">Infiriendo sobre la proporción</font>

Consideraremos ahora los datos para inferir sobre la proporción de pingüinos machos sobre el total de la población

La inferencia sobre una proporción nos permite estimar la proporción de una característica (en este caso, el porcentaje de pingüinos machos) en una población a partir de los datos de una muestra. Esto se hace calculando un intervalo de confianza para la proporción.

* Proporción de una característica: Es el porcentaje de elementos en una muestra que tienen una cierta característica. Por ejemplo, si tenemos una muestra de 100 pingüinos Adelia y 40 de ellos son machos, la proporción de pingüinos machos sería 40%


* Intervalo de confianza para la proporción: Similar a los intervalos de confianza para la media, un intervalo de confianza para una proporción nos dice en qué rango se espera que esté la verdadera proporción en la población, con un nivel de confianza determinado (por ejemplo, 95%).

La inferencia sobre la proporción se utiliza para estimar la proporción de una característica en una población a partir de una muestra. En este caso, estamos interesados en la proporción de pingüinos machos en la especie Adelia.

**Esto les servirá para el ejercicio 2 del desafío**

```python
import pandas as pd
import seaborn as sns
import statsmodels.api as sm

# Cargar el conjunto de datos de pingüinos desde Seaborn
pinguins = sns.load_dataset("penguins")

# Filtrar los datos para obtener solo los pingüinos Adelia
adelia = pinguins[pinguins["species"] == "Adelie"]

# Tamaño de la muestra
n = len(adelia)

#tips
#si te pidieran sacar 50 ejemplos al azar, sería:

#muestra = np.random.choice(adelia, 50)

# Contar la cantidad de machos en la muestra
num_machos = adelia[adelia["sex"] == "Male"]["sex"].count()

# Proporción muestral de machos en la muestra
proporcion_muestral = num_machos / n

# Nivel de confianza (por ejemplo, 95%)
confianza = 0.95

# Calcular el intervalo de confianza utilizando statsmodels
intervalo_confianza = sm.stats.proportion_confint(num_machos, n, alpha=1 - confianza)

# Imprimir el resultado
print(f"Intervalo de confianza del {confianza * 100}% para la proporción de machos en la población de pingüinos Adelia:")
print(f"({intervalo_confianza[0]}, {intervalo_confianza[1]})")

```

En la línea, donde se utiliza la función **proportion_confint** de **statsmodels** para calcular el intervalo de confianza para la proporción de machos. El argumento alpha es el nivel de significancia, que se establece como **1−confianza**. Por ejemplo, para un nivel de confianza del 95%, 
𝛼
α sería 
0.05.

### Interpretación del Intervalo de Confianza

**Interpretación:** El intervalo de confianza del 95% para la proporción de machos en la población de pingüinos Adelia es aproximadamente (0.40, 0.56). Esto significa que estás 95% seguro de que la verdadera proporción de machos en la población de pingüinos Adelia se encuentra entre 40.08% y 55.97%.

**Conclusión:** Si este intervalo incluye el valor que estás evaluando (por ejemplo, si quieres ver si al menos el 50% de los pingüinos Adelia son machos), puedes usar esta información para hacer inferencias sobre la población general. Si decides hacer una hipótesis, puedes establecerla con base en este intervalo. Si el límite inferior es mayor a la proporción que consideras relevante (como el 50%), puedes rechazar la hipótesis nula de que la proporción es menor o igual al 50%.

## <font color="green">Prueba de hipótesis para muestras independientes</font>

Compararemos ahora la diferencia en el peso corporal entre las especies de pingüinos "Adelie" y "Chinstrap". La hipótesis nula (H0) sería que no hay diferencia significativa en el peso corporal entre las especies, y la hipótesis alternativa (H1) es que sí hay una diferencia significativa.


Para realizar una prueba de hipótesis para muestras independientes y comparar el peso corporal entre las especies de pingüinos "Adelie" y "Chinstrap", podemos realizar un t-test de dos muestras independientes, ya que estamos tratando de comparar las medias de dos grupos distintos.

### Pasos para realizar la prueba de hipótesis:

#### Formulación de hipótesis:

**Hipótesis nula (H₀):** No hay diferencia significativa en el peso corporal entre las especies "Adelie" y "Chinstrap".

**Hipótesis alternativa (H₁):** Hay una diferencia significativa en el peso corporal entre las especies "Adelie" y "Chinstrap".


### Carga datos
**Cargar y preparar los datos:** Utilizaremos el dataset de penguins de Seaborn y filtraremos las dos especies que queremos comparar ("Adelie" y "Chinstrap").

**Realizar la prueba t de muestras independientes:** Verificaremos si hay diferencia significativa entre las medias de peso corporal de las dos especies.

**Esto les servirá para el ejercicio 3 del desafío**

```python
import pandas as pd
import seaborn as sns
from scipy import stats

# Cargar el conjunto de datos de pingüinos desde Seaborn
pinguins = sns.load_dataset("penguins")

# Filtrar los datos para obtener los pesos corporales de las especies "Adelie" y "Chinstrap"
adelie = pinguins[pinguins["species"] == "Adelie"]["body_mass_g"]
chinstrap = pinguins[pinguins["species"] == "Chinstrap"]["body_mass_g"]

# Eliminar filas con valores nulos en los datos de peso corporal
adelie = adelie.dropna()
chinstrap = chinstrap.dropna()

# Realizar una prueba de hipótesis (prueba t de muestras independientes)
t_stat, p_value = stats.ttest_ind(adelie, chinstrap, equal_var=False)

# Nivel de significancia (alfa)
alpha = 0.05

# Comprobar si el valor p es menor que alfa para rechazar la hipótesis nula
if p_value < alpha:
    print("Resultado: Se rechaza la hipótesis nula")
else:
    print("Resultado: No se rechaza la hipótesis nula")

# Imprimir el valor t y el valor p
print(f"Valor t: {t_stat}")
print(f"Valor p: {p_value}")

```

## Explicación de los pasos:

**Cargar los datos:** Utilizamos seaborn para cargar el dataset de pingüinos.

**Filtrar las especies:** Seleccionamos solo los pingüinos "Adelie" y "Chinstrap" y sus pesos corporales (body_mass_g).

**Prueba t de dos muestras:** La función **ttest_ind** de **scipy.stats** realiza una prueba t de Student para comparar las medias de dos muestras independientes.

**Interpretación:** Si el valor p es menor que el nivel de significancia (usualmente 0.05), rechazamos la hipótesis nula y concluimos que hay una diferencia significativa.

### Resultados posibles:

**Si p < 0.05:** Hay evidencia suficiente para rechazar la hipótesis nula y concluir que existe una diferencia significativa en el peso corporal entre ambas especies.


**Si p ≥ 0.05:** No hay suficiente evidencia para rechazar la hipótesis nula, lo que indicaría que no hay una diferencia significativa en el peso corporal entre las especies.

Este enfoque es ideal para comparar las medias de dos grupos independientes y determinar si existe una diferencia estadísticamente significativa.

* Cuando ejecutas una prueba t de Student, como en el ejemplo que estamos viendo, se obtienen dos valores principales: 
**el valor t (t_stat)** y **el valor p (p_value)**. Ambos son necesarios para interpretar el resultado de la prueba estadística.

## ¿Por qué se obtienen juntos?

* La razón es que el valor t y el valor p están estrechamente relacionados, pero proporcionan diferentes formas de interpretar los resultados:

**El valor t (t_stat):**

Es una medida de la magnitud de la diferencia entre la media de los datos observados y la media hipotética, ajustada por la variabilidad de los datos. Cuanto mayor sea este valor, más improbable es que la diferencia observada sea debida al azar.
El valor t por sí solo te da una idea de cuán "lejos" está la media observada de la media hipotética, en unidades de desviación estándar.

**El valor p (p_value):**

Es la probabilidad de observar un valor t tan extremo (o más extremo) que el calculado, bajo la suposición de que la hipótesis nula es verdadera. Es una medida más directa de significancia estadística.
Mientras que el valor t te da una medida cuantitativa de la diferencia, el valor p te dice la probabilidad de que esa diferencia ocurra por casualidad.
En resumen:

* El valor t te dice cuán grande es la diferencia relativa.

* El valor p te dice si esa diferencia es significativa estadísticamente (normalmente se compara con un nivel de significancia como 0.05).


Por eso, se devuelven juntos: necesitas el valor t para entender la magnitud de la diferencia, y el valor p para saber si es estadísticamente significativa.
