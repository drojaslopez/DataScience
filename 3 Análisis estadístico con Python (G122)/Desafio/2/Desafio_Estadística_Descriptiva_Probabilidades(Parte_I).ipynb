En este notebook realizaremos un análisis estadístico descriptivo sobre salarios en Data Science y luego aplicaremos conceptos de probabilidad utilizando datos de la Copa del Mundo 2014.

```python
import pandas as pd

# Cargar el dataset de salarios
df = pd.read_csv("worldcup2014.csv")
total = len(df)
total
```

```python
df.head(5)
#df.head(total)
```

```python
condicionAfrica = df['continent'] == 'africa'
condicionAsia = df['continent'] == 'asia'
condicionEuropa = df['continent'] == 'europe'
condicionNorteamerica = df['continent'] == 'northamerica'
condicionSudamerica = df['continent'] == 'southamerica'
condicionClasificado = df['clasificado'] == 1
condicionGanados = df['juegos_ganados'] > 0


#print(condicionAfrica)
```

```python
#pertenecer a África
p_Africa=condicionAfrica.sum() / total
print(f"pertenecer a África: {p_Africa:.3%}")

#pertenecer a Asia
p_Asia = condicionAsia.sum() / total
print(f"pertenecer a Asia: {p_Asia:.3%}")

#pertenecer a Europa
p_Europa = condicionEuropa.sum() / total
print(f"pertenecer a Europa: {p_Europa:.3%}")

#pertenecer a Norteamérica
p_Norteamerica= condicionNorteamerica.sum() / total
print(f"pertenecer a Norteamérica: {p_Norteamerica:.3%}")

#pertenecer a Sudamérica
p_Sudamerica = condicionSudamerica.sum() / total
print(f"pertenecer a Sudamérica: {p_Sudamerica:.3%}")

#clasiﬁcar a segunda ronda
p_Clasificado=condicionClasificado.sum() / total
print(f"Clasiﬁcar a segunda ronda: {p_Clasificado:.3%}")

#ganar al menos un partido
p_Ganados=condicionGanados.sum() / total
print(f"Ganar al menos un partido: {p_Ganados:.3%}")

#pertenecer a africa y clasiﬁcar a segunda ronda
p_AfricaClasiﬁcar=(condicionAfrica & condicionClasificado).sum() / total
print(f"pertenecer a africa y clasiﬁcar a segunda ronda: {p_AfricaClasiﬁcar:.3%}")

#pertenecer a Europa y clasiﬁcar a segunda ronda
p_EuropaClasiﬁcar=(condicionEuropa & condicionClasificado).sum() / total
print(f"pertenecer a Europa y clasiﬁcar a segunda ronda: {p_EuropaClasiﬁcar:.3%} ")

#pertenecer a Asia y clasiﬁcar a segunda ronda
p_AsiaClasiﬁcar=(condicionAsia & condicionClasificado).sum() / total
print(f"pertenecer a Asia y clasiﬁcar a segunda ronda: {p_AsiaClasiﬁcar:.3%} ")

#pertenecer a Sudamérica y clasiﬁcar a segunda ronda
p_SudamericaClasiﬁcar=(condicionSudamerica & condicionClasificado).sum() / total
print(f"pertenecer a Sudamérica y clasiﬁcar a segunda ronda: {p_SudamericaClasiﬁcar:.3%} ")

#pertenecer a Norteamérica y clasiﬁcar a segunda ronda
p_NorteamericaClasiﬁcar=(condicionNorteamerica & condicionClasificado).sum() / total
print(f"pertenecer a Norteamérica y clasiﬁcar a segunda ronda: {p_EuropaClasiﬁcar:.3%} ")


#pertenecer a África o clasiﬁcar a segunda ronda
p_AfricaoClasiﬁcar=(condicionAfrica | condicionClasificado).sum() / total
print(f"pertenecer a África o clasiﬁcar a segunda ronda: {p_AfricaoClasiﬁcar:.3%} ")

#pertenecer a Europa o clasiﬁcar a segunda ronda
p_EuropaoClasiﬁcar=(condicionEuropa | condicionClasificado).sum() / total
print(f"pertenecer a Europa o clasiﬁcar a segunda ronda: {p_EuropaoClasiﬁcar:.3%} ")

#pertenecer a Asia o clasiﬁcar a segunda ronda
p_AsiaoClasiﬁcar=(condicionAsia | condicionClasificado).sum() / total
print(f"pertenecer a Asia o clasiﬁcar a segunda ronda: {p_AsiaoClasiﬁcar:.3%} ")


#pertenecer a Sudamérica o clasiﬁcar a segunda ronda
p_SudamericaoClasiﬁcar=(condicionSudamerica | condicionClasificado).sum() / total
print(f"pertenecer a Sudamérica o clasiﬁcar a segunda ronda: {p_SudamericaoClasiﬁcar:.3%} ")

#pertenecer a Norteamérica o clasiﬁcar a segunda ronda
p_NorteamericaoClasiﬁcar=(condicionNorteamerica | condicionClasificado).sum() / total
print(f"pertenecer a Norteamérica o clasiﬁcar a segunda ronda: {p_NorteamericaoClasiﬁcar:.3%} ")

#pertenecer a África y haber ganado al menos un partido
p_AfricaGanar=(condicionAfrica & condicionGanados).sum() / total
print(f"pertenecer a África y ganado al menos un partido: {p_AfricaGanar:.3%} ")


#pertenecer a Europa y haber ganado al menos un partido
p_EuropaGanar=(condicionEuropa & condicionGanados).sum() / total
print(f"pertenecer a Europa y ganado al menos un partido: {p_EuropaGanar:.3%} ")


#pertenecer a Asia y haber ganado al menos un partido
p_AsiaGanar=(condicionAsia & condicionGanados).sum() / total
print(f"pertenecer a Asia y ganado al menos un partido: {p_AsiaGanar:.3%} ")


#pertenecer a Sudamérica y haber ganado al menos un partido
p_SudamericaGanar=(condicionSudamerica & condicionGanados).sum() / total
print(f"pertenecer a Sudamérica y ganado al menos un partido: {p_SudamericaGanar:.3%} ")

#pertenecer a Norteamérica y haber ganado al menos un partido
p_NorteamericaGanar=(condicionNorteamerica & condicionGanados).sum() / total
print(f"pertenecer a Norteamérica y ganado al menos un partido: {p_NorteamericaGanar:.3%} ")

#haber ganado al menos un partido y clasiﬁcar a segunda ronda
p_GanadoClasificado=(condicionGanados & condicionClasificado).sum() / total
print(f"haber ganado al menos un partido y clasiﬁcar a segunda ronda: {p_GanadoClasificado:.3%} ")
```

```python
#pertenecer a Europa, si se sabe que clasificó
p_Europa_Dado_Clasificado = p_EuropaClasificar / p_Clasificado
print(f"pertenecer a Europa, si se sabe que clasificó: {p_Europa_Dado_Clasificado:.3%} ")

#pertenecer a Sudamérica, si se sabe que clasificó
p_Sudamerica_Dado_Clasificado = p_SudamericaClasificar / p_Clasificado
print(f"pertenecer a Sudamérica, si se sabe que clasificó: {p_Sudamerica_Dado_Clasificado:.3%} ")

#pertenecer a África, si se sabe que ganó al menos un partido

p_Africa_Dado_Ganado = p_AfricaGanar / p_Ganados
print(f"pertenecer a África, si se sabe que ganó al menos un partido: {p_Africa_Dado_Ganado:.3%} ")

#haber ganado al menos un partido, si se sabe que clasificó

p_Ganado_Dado_Clasificado = p_GanadoClasificado / p_Clasificado
print(f"haber ganado al menos un partido, si se sabe que clasificó: {p_Ganado_Dado_Clasificado:.3%} ")

#haber ganado al menos un partido, si se sabe que pertenece a Norteamérica

p_Ganado_Dado_Norteamerica = p_NorteamericaGanar / p_Norteamerica
print(f"haber ganado al menos un partido, si se sabe que pertenece a Norteamérica: {p_Ganado_Dado_Norteamerica:.3%} ")

#NO haber clasificado, si se sabe que pertenece a África

p_NoClasificado_Dado_Africa = 1 - (p_AfricaClasificar / p_Africa)
print(f"NO haber clasificado, si se sabe que pertenece a África: {p_NoClasificado_Dado_Africa:.3%} ")
```
