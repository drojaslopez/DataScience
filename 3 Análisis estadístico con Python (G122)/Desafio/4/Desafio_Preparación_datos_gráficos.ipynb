### 2. Considera los pesos de los bebés. ¿Se puede decir que siguen una distribución normal? Justifica tu respuesta creando los gráficos y modelos necesarios.


```python
df_peso_bebe = df[['Birth.Weight']].copy()
df_peso_bebe.describe()
#calcular la distribucion normal para la columna birth weight
df_peso_bebe = df_peso_bebe.dropna()
mean = df_peso_bebe['Birth.Weight'].mean()
std = df_peso_bebe['Birth.Weight'].std()


```

```python
# 1. Create histogram with normal distribution overlay

plt.figure(figsize=(12, 6))
sns.histplot(df_peso_bebe['Birth.Weight'], kde=True, stat='density', alpha=0.7)
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, mean, std)  # Ahora sí funcionará
plt.plot(x, p, 'k', linewidth=2, label=f'Normal (μ={mean:.1f}, σ={std:.1f})')
plt.legend()
plt.title('Distribución de pesos de bebés vs Distribución Normal')
plt.show()
```

### 4. Compara los pesos de los bebés de madres fumadoras y no fumadoras. ¿Qué diferencias existen? Explica y construye los modelos correspondientes.

```python
df_fumadoras = df[['Maternal.Smoker', 'Birth.Weight']].copy()
```

```python
bebe_fumadores = df_fumadoras[df_fumadoras['Maternal.Smoker'] == True]['Birth.Weight']


bebe_no_fumadores =df_fumadoras[df_fumadoras['Maternal.Smoker']==False]['Birth.Weight']
```

```python
bebe_fumadores
```

```python
bebe_no_fumadores
```

```python
def get_pdf(x):
  mu=x.mean()
  sigma=x.std()
  minimo=x.min()
  maximo=x.max()
  x = np.linspace(minimo, maximo)
  pdf = norm.pdf(x, loc=mu, scale=sigma)
  return(x,pdf)
```

```python
x1, pdf = get_pdf(bebe_fumadores)
plt.plot(x1 , pdf, color='red', label='bebe_fumadores')

x2,pdf=get_pdf(bebe_no_fumadores)
plt.plot(x2 , pdf, color='blue', label='bebe_no_fumadores')

plt.hist(bebe_fumadores, bins=100, density=True, alpha=0.2, color='red')
plt.hist(bebe_no_fumadores, bins=100, density=True, alpha=0.2, color='blue')

plt.xlabel('x')
plt.ylabel('Densidad de probabilidad normalizada')
plt.title('Comparación de pesos de bebe_fumadores y bebe_no_fumadores')
plt.legend()
```
