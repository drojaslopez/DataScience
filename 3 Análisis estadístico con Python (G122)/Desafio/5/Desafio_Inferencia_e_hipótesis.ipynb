```python
# Datos reales de fumadoras (convertir a numérico: True=1, False=0)
datos_reales = df['Maternal.Smoker'].astype(int).values

# Parámetros
n_muestra = 8  # tamaño de cada muestra
n_simulaciones = 1000  # número de muestras

# Extraer 1000 muestras aleatorias de tamaño 8 de los datos reales
medias_muestrales_reales = []

for i in range(n_simulaciones):
    # Muestreo aleatorio sin reemplazo
    muestra = np.random.choice(datos_reales, size=n_muestra, replace=False)
    media_muestral = np.mean(muestra)
    medias_muestrales_reales.append(media_muestral)

medias_muestrales_reales = np.array(medias_muestrales_reales)

# Mostrar resultados
print(f"Se extrajeron {n_simulaciones} muestras de tamaño {n_muestra} de los datos reales")
print(f"Total de datos disponibles: {len(datos_reales)}")
print(f"Proporción real de fumadoras: {datos_reales.mean():.4f}")
print(f"Primeras 10 medias muestrales: {medias_muestrales_reales[:10]}")
print(f"Media de las medias muestrales: {medias_muestrales_reales.mean():.4f}")
print(f"Desviación estándar de las medias: {medias_muestrales_reales.std():.4f}")
```
